```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class LinkedList:
    def __init__(self, value):
        length = 0
        self.length = length
        new_node = Node(value)
        self.head = new_node
        self.tail = new_node
        self.length += 1

    def norm_append(self, value):
        g_node = Node(value)
        self.tail.next = g_node
        self.tail = g_node
        self.length +=1
        return self.head

    def rolling_append(self, value):
        for i in range(value):
            n_node = Node(i)
            self.tail.next = n_node
            self.tail = n_node
            self.length += 1
        return self.head

    def norm_prepend(self,value):
        gr_node = Node(value)
        gr_node.next = self.head
        self.head = gr_node
        self.length += 1

    def rolling_prepend(self, value):
        for i in range(value,0,-1):
            i = Node(i)
            i.next = self.head
            self.head = i
            self.length += 1

    def pop(self):
        popper = self.head
        if self.length>2:
            while popper.next.next is not None:
                popper = popper.next
                if popper.next.next is None:
                    break
        elif self.length == 2:
            popper = self.head
        pot = self.tail
        self.tail = popper
        self.tail.next = None
        self.length -= 1
        return pot.value

    def pop_first(self):
        if self.length >= 2:
            tot = self.head.next
            popper = self.head
            self.head = tot
            self.length -= 1
            return popper.value
        else:
            tep = self.head
            self.head = None
            self.tail = None
            self.length-=1
            return tep

    def get(self, index):
        getter = self.head
        if index < 0 or index >= self.length:
            return f'LOLzzz'
        else:
            for _ in range(index):
                getter = getter.next
            return getter.value

    def set_value(self, index, value):
        getter = self.head
        if index<0 or index>=self.length:
            return f'LOLzzz'
        else:
            for _ in range(index):
                getter = getter.next
            getter.value = value

    def insert(self, index, value):
        bn_node = Node(value)
        insrt = self.head
        botl = self.head
        if index<0 or index >= self.length:
            return f'LOLzzz'
        else:
            for i in range(index):
                insrt = insrt.next
            for i in range(index-1):
                botl = botl.next
            botl.next = bn_node
            bn_node.next = insrt
            self.length += 1

    def remove(self, index):
        insrt = self.head
        botl = self.head
        if index<0 or index >= self.length:
            return f'LOLzzz'
        else:
            for i in range(index):
                insrt = insrt.next
            for i in range(index-1):
                botl = botl.next
            botl.next = insrt.next
            insrt.next = None
            self.length -= 1
            return insrt.value

    def reverse(self):
        temp = self.head
        self.head = self.tail
        self.tail = temp
        after = temp.next
        before = None
        for _ in range(self.length):
            after = temp.next
            temp.next = before
            before = temp
            temp = after

    def printDetails(self):
        temp = self.head
        while temp is not None:
            print(f'{temp.value}==>',end='')
            temp = temp.next
        print('None')

bot = LinkedList(3)
bot.rolling_append(8)
print(bot.head.value)
print(bot.head.next.next.next.value)
print(bot.tail.value)
bot.printDetails()
print(bot.pop())
bot.norm_append(4)
bot.printDetails()
bot.norm_prepend(5)
bot.printDetails()
bot.rolling_prepend(6)
bot.printDetails()
print(bot.pop_first())
bot.printDetails()
print(bot.get(6))
bot.set_value(6,90)
bot.printDetails()
bot.insert(3, 69)
bot.printDetails()
print(bot.remove(9))
bot.printDetails()
print(bot.length)
bot.reverse()
bot.printDetails()
```
