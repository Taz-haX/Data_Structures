```python
class Node:
    def __init__(self, value):
        next = None
        prev = None
        self.value = value
        self.next = next
        self.prev = prev

class DoublyLinkedList:
    def __init__(self, value):
        self.length = 0
        new_node = Node(value)
        self.head = new_node
        self.tail = new_node
        self.length += 1

    def norm_append(self, value):
        n_node = Node(value)
        self.tail.next = n_node
        n_node.prev = self.tail
        self.tail = n_node
        self.length += 1

    def rolling_append(self, value):
        for i in range(value):
            bn_node = Node(i)
            self.tail.next = bn_node
            bn_node.prev = self.tail
            self.tail = bn_node
            self.length += 1

    def pop(self):
        popper = self.tail
        self.tail = popper.prev
        self.tail.next = None
        popper.prev = None
        self.length -= 1
        return popper.value

    def prepend(self, value):
        n_node = Node(value)
        self.head.prev = n_node
        n_node.next = self.head
        self.head = n_node
        self.length += 1

    def pop_first(self):
        popper = self.head
        self.head = popper.next
        self.head.prev = None
        popper.next = None
        self.length -= 1
        return popper.value

    def get(self, index):
        getter = self.head
        for _ in range(index):
            getter = getter.next
        bok = getter.next
        hok = getter.prev
        hok.next = bok
        bok.prev = hok
        getter.next = None
        getter.prev = None
        self.length -= 1
        print(getter.value)

    def set_value(self, index, value):
        setter = self.head
        for _ in range(index):
            setter = setter.next
        setter.value = value

    def insert(self, index, value):
        inserter = self.head
        for _ in range(index):
            inserter = inserter.next
        con = inserter.prev
        gn_node = Node(value)
        gn_node.prev = con
        gn_node.next = inserter
        inserter.prev = gn_node
        con.next = gn_node
        self.length += 1

    def remove(self, index):
        remover = self.head
        for _ in range(index):
            remover = remover.next
        con = remover.prev
        bon = remover.next
        con.next = bon
        bon.prev = con
        remover.next = None
        remover.prev = None
        self.length -= 1
        return remover.value

    def showList(self):
        temp = self.head
        print(f'None<==>', end='')
        while temp is not None:
            print(f'{temp.value}<==>', end='')
            temp = temp.next
        print(f'None', end='')
        print(f' [Total Elements: {self.length}]')

bl = DoublyLinkedList(4)
bl.rolling_append(7)
bl.norm_append(9)
bl.showList()
print(bl.pop())
bl.showList()
bl.prepend(44)
bl.showList()
print(bl.pop_first())
bl.showList()
bl.get(4)
bl.showList()
bl.set_value(5,69)
bl.showList()
bl.insert(2,23)
bl.showList()
print(bl.remove(5))
bl.showList()
```
